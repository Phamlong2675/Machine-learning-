```python
# Change hàm ra quyết định
# --- Rule-based: g(s) = 1 nếu pos_freq > neg_freq, ngược lại 0 ---

def predict_tweet_rule(tweet, freqs, tokenizer, normalize=False, train_len=None):
    tokens = tokenizer(tweet)

    # Đếm tần suất tích cực/tiêu cực từ từ điển freqs: {(word, label): count}
    pos = sum(freqs.get((w, 1.0), 0) for w in tokens)
    neg = sum(freqs.get((w, 0.0), 0) for w in tokens)

    # (Tuỳ chọn) Chuẩn hoá theo mục 3 của bài: N = train_len * độ_dài_câu
    if normalize:
        if train_len is None:
            raise ValueError("Cần truyền train_len khi normalize=True")
        N = max(1, len(tokens)) * train_len
        pos = pos / N
        neg = neg / N

    # Luật tay g(s)
    return 1 if pos > neg else 0  # hoà thì quy ước 0

# ------------------ Test results (giữ khung của bạn) ------------------
my_tweet = 'Hopefully my very first NLP session is not too boring to you " !!!'
print(process_tweet(my_tweet))

# DÙNG luật tay thay vì y_hat xác suất của Logistic
y_hat = predict_tweet_rule(my_tweet, freqs, process_tweet, normalize=False)
print(y_hat)

if y_hat == 1:
    print('Positive sentiment')
else:
    print('Negative sentiment')
```
